```python
def weighted_absolute_error(y_pred,y_true):
    points=0
    for i in range(len(y_true)):
        error=abs(y_pred[i]-y_true[i])
        
        if 0<=error<=0.05:
            points+=1
        
        elif 0.05<error<=0.1:
            points+=0.5
        
        elif 0.1<error<=0.5:
            points+=0.25
        
        else:
            points+=0
    
    return points
```

```python
def duration_error(y_pred,y_true, bound=0.1):
    count=0
    ghost=0
    points=0
    for i in range(len(y_true)):
        error = abs(y_pred[i]-y_true[i])
        if error>0.1:
            count+=1
            ghost+=1
        else:
            count=0
        
        if count==0 and ghost==0:
            points+=1
            ghost=0
        
        elif count==0 and ghost==1:
            points+=0.5
            ghost=0
        
        elif count==0 and 2<=ghost<=10:
            points+=0.25
            ghost=0
        
        else:
            points+=0
            ghost=0

    return points
        
```
